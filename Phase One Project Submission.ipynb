```python
#changing the dataframe name
Aviation_data = df

print(Aviation_data)
```

```python
# Replacing missing values with the median
Aviation_data_columns = ['Number.of.Engines', 'Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries', 'Total.Uninjured']
Aviation_data[Aviation_data_columns] = Aviation_data[Aviation_data_columns].fillna(Aviation_data[Aviation_data_columns].median())
Aviation_data
```

```python
# Replacement of all missing categorical values
Aviation_data.fillna('Missing', inplace= True)
Aviation_data
```

```python
# Updating the CSV file after data cleaning
Aviation_data=Aviation_data.applymap(lambda x: x.encode('unicode_escape').decode('utf-8') if isinstance(x, str)else x)
```

```python
# Dropping duplicates
Aviation_data.drop_duplicates()
Aviation_data
```

Data Manipulation

I used the groupby function to summarise the data.

```python
# Grouping by Aircraft category and calculate mean number of accidents
df = Aviation_data.groupby('Aircraft.Category').agg({
    'Total.Fatal.Injuries': 'mean',
    'Total.Serious.Injuries': 'mean',
    'Total.Minor.Injuries': 'mean',
    'Total.Uninjured': 'mean'
}).reset_index()
df
```
